```python
from pyecharts.globals import CurrentConfig,NotebookType
CurrentConfig.NOTEBOOK_TYPE=NotebookType.JUPYTER_NOTEBOOK
from pyecharts.charts import Bar
from pyecharts import options as opts
import pymysql

db = pymysql.connect(host='localhost', user='root',password='Passwd1!',port=3306,db='spark')
cursor = db.cursor()
sql = 'select * from Amazon_ReviewsTop15 limit 15'
cursor.execute(sql)
data = cursor.fetchall()
x, y =[], []
for i in data:
    x.append(i[0])
    y.append(i[1])    

def bar() -> Bar:
    c = (
    Bar()
        .add_xaxis(x)
        .add_yaxis("数量", y,category_gap="50%",gap='180%')   # category_gap柱体宽度
     #  .reversal_axis()      # 旋转坐标轴
        .set_global_opts(
            opts.TitleOpts(title="图书数量分析",subtitle="前15图书数据",pos_left='center'),
            toolbox_opts=opts.ToolboxOpts(),
            legend_opts=opts.LegendOpts(is_show=True,pos_left='25%'),       
            xaxis_opts=opts.AxisOpts(name="图书名称",axislabel_opts={"interval":0,"rotate":0}), 
            yaxis_opts=opts.AxisOpts(name="数量"),
            datazoom_opts=opts.DataZoomOpts()    #设置滑块
            )
            .set_series_opts(label_opts=opts.LabelOpts(is_show=False),
                    markpoint_opts=opts.MarkPointOpts(
                        data=[
                            opts.MarkLineItem(type_="min", name="最小值"),
                            opts.MarkLineItem(type_="max", name="最大值"),
                            opts.MarkLineItem(type_="average", name="平均值")])
                    
    )
        
    )
    return c

bar().load_javascript()
bar().render("条形图.html")
bar().render_notebook()
```

```python
from pyecharts.globals import *
CurrentConfig.NOTEBOOK_TYPE=NotebookType.JUPYTER_NOTEBOOK
from pyecharts import options as opts
from pyecharts.charts import *
from pymysql import connect
db = connect(host='localhost', user='root',password='Passwd1!',port=3306,db='spark')
cursor = db.cursor()
sql = 'select * from PriceCounts'
cursor.execute(sql)
data = cursor.fetchall()
x, y =[], []
for i in data:
    x.append(i[0])
    y.append(i[1])
y1 = y

c = (
    Bar()
    .add_xaxis(x)
    .add_yaxis("数 量",y,label_opts = opts.LabelOpts(is_show=False))
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(name="价格",axislabel_opts={"interval":0}),
        yaxis_opts=opts.AxisOpts(name="数量"),
        title_opts=opts.TitleOpts(title="书籍价格及数量分析")
    )
    .extend_axis(
    yaxis=opts.AxisOpts(name="数量",
          axisline_opts=opts.AxisLineOpts(is_show=True,             
          linestyle_opts=opts.LineStyleOpts()), 
         )
    )
)
line = Line().add_xaxis(x).add_yaxis("数量",y,yaxis_index=1,is_smooth=True).set_series_opts(linestyle_opts=opts.LineStyleOpts(width=2))
c.overlap(line)


c.render_notebook()
```

##### 双条形+双折线

```python
db = connect(host='localhost',user='root',passwd='Passwd1!',port=3306,db='spark')
cursor = db.cursor()
sql = "select * from YearGen1"
cursor.execute(sql)
data = cursor.fetchall()
x,data1,data2=[],[],[]
y1,y2=[],[]
n=0
for i in data:
    data1.append(i[1])
    data2.append(i[2])
    if (data1[n][0]=='N'):
        x.append(i[0])
        y1.append(i[2])
    else:
        y2.append(i[2])
    n+=1  
x = [str(i) for i in x]   
c=(
Bar()
    .add_xaxis(x)
    .add_yaxis("Non Fiction" ,y1,z=0,color="pink")
    .add_yaxis("Fiction", y2,z=0,color="green")
    .set_global_opts(
        title_opts=opts.TitleOpts(title="条形图",subtitle="随着年份的变化不同类别的书籍数量对比"),
        
        )
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))   #柱状图的标签数值隐藏

    .extend_axis(
        yaxis=opts.AxisOpts(
        is_show=False,
        )
    )
)
line = (
    Line()
    .add_xaxis(x)
    .add_yaxis("Non Fiction",y1,yaxis_index=1,is_smooth=True)
    .add_yaxis("Fiction",y2,yaxis_index=1,is_smooth=True)
    .set_series_opts(linestyle_opts=opts.LineStyleOpts(width=5))
)

c.overlap(line)
c.render_notebook()
```
