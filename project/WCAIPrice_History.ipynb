```python
#使用Spark读取csv文件
from pyspark import SparkConf, SparkContext
from pyspark.sql import SparkSession, Row
from pyspark.sql.types import *

sc = SparkContext(conf= SparkConf().setMaster('local'))
spark = SparkSession.builder.getOrCreate()

sparkdf1 = spark.read.options(header='true', inferSchema='True', delimiter=',').csv('data/cities_part1.csv')
sparkdf2 = spark.read.options(header='true', inferSchema='True', delimiter=',').csv('data/cities_part2.csv')
```

```python
#两表合并
sparkdf = sparkdf1.unionAll(sparkdf2)
sparkdf.show()
```

## 分析上榜城市所属国家的数量

```python
CountryCountsRDD = sparkdf.select('Country').rdd.map(lambda line: (line['Country'], 1))\
                .reduceByKey(lambda x,y: x+y)
CountryCountsDF = CountryCountsRDD.toDF().withColumnRenamed('_1', 'Country')\
                .withColumnRenamed('_2', 'Counts')
CountryCountsDF = CountryCountsDF.sort(CountryCountsDF['Counts'].desc())

prop = {}
prop['user'] = 'root'
prop['password'] = 'Passwd1!'
prop['driver'] = 'com.mysql.jdbc.Driver'

CountryCountsDF.write.jdbc('jdbc:mysql://localhost:3306/spark', 'CountryCounts', 'append', prop)
```
